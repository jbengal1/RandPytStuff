```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
import matplotlib.animation as PillowWriter
from itertools import combinations
```

```python
x = np.linspace(-10, 10, 1000)
y = 2 + 3*x + x**2 + 0.1*x**3
```

```python
plt.plot(x, y)
```

```python
vec = np.array([x, y])
```

```python
np.shape(vec)
```

```python
new_vecs = []
thetas = [np.radians(n*10) for n in range(360)]
for theta in thetas:
    c, s = np.cos(theta), np.sin(theta)
    R = np.array(((c, -s), (s, c)))
    new_vecs.append(np.dot(R, vec))
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,5))


def animate(i):
    ax.clear()
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    ax.plot(new_vecs[i][0], new_vecs[i][1])

    
ani = animation.FuncAnimation(fig, animate, frames=int(36), interval=50)
file_name = 'parabRotate.gif'
ani.save(file_name,writer='pillow',fps=15,dpi=100)
```
